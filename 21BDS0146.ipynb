```python
# Module 2 : Data Transformation
```

```python
# Data Deduplication
```

```python
duplicate_rows = data[data.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
```

```python
data_deduplicated = data.drop_duplicates()
print("Shape of data after deduplication:", data_deduplicated.shape)
```

```python
# Replacing Values
```

```python
median_murders = data['murders'].median()
data['murders'] = data['murders'].replace(0, median_murders)
data[['murders']].head()
```

```python
# Discretization and Binning
```

```python
bins = [0, 1, 2, data['murdrate'].max()]
labels = ['Low', 'Medium', 'High']
data['murdrate_bin'] = pd.cut(data['murdrate'], bins=bins, labels=labels, include_lowest=True)
data[['murdrate', 'murdrate_bin']].head()
```

```python
# Missing Data Handling
```

```python
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
data['arrests'] = data['arrests'].fillna(data['arrests'].mean())
```
