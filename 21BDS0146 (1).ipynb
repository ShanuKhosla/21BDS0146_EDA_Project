```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/salemprakash/EDA/refs/heads/main/Data/countymurders.csv'
data = pd.read_csv(url)
```

```python
data.fillna(data.mean(), inplace = True)
```

```python
#Univariate Analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (8, 6))
sns.histplot(data['murders'], kde=True)
plt.title('Distribution of Murders')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['arrests'], kde=True)
plt.title('Distribution of Arrests')
plt.show()
```

```python
data['statefips'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['murdrate'])
plt.title('Box Plot of Murder Rate')
plt.show()
```

```python
#Bivariate Analysis
```

```python
correlation_matrix = data.corr()
```

```python
correlation_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='murders', y='arrests', data = data)
plt.title("ScatterPlot of Murders v/s Arrests")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='statefips', y='murdrate', data=data)
plt.title('Murder Rate by State')
plt.show()
```

```python
#Multivariate Analysis
```

```python
sns.pairplot(data[['murders', 'arrests', 'murdrate', 'arrestrate']])
plt.suptitle('Pairplot', y=1.02)
plt.show()
```
